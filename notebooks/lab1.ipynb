# –õ–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–∞—è —Ä–∞–±–æ—Ç–∞ ‚Ññ1: –í–≤–µ–¥–µ–Ω–∏–µ –≤ GigaChat

**–¶–µ–ª—å:** –ü–æ–∑–Ω–∞–∫–æ–º–∏—Ç—å—Å—è —Å GigaChat API, –≤—ã–ø–æ–ª–Ω–∏—Ç—å –±–∞–∑–æ–≤—ã–µ –∑–∞–ø—Ä–æ—Å—ã, –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –æ—Ç–≤–µ—Ç—ã –∏ –≤—ã—è–≤–∏—Ç—å ¬´–≥–∞–ª–ª—é—Ü–∏–Ω–∞—Ü–∏–∏¬ª.

–ò—Å–ø–æ–ª—å–∑—É–µ–º –º–æ–¥–µ–ª—å: **Gigachat-2-Pro**

–í—Å–µ –∑–∞–ø—Ä–æ—Å—ã –≤—ã–ø–æ–ª–Ω—è—é—Ç—Å—è —á–µ—Ä–µ–∑ –±–µ–∑–æ–ø–∞—Å–Ω—ã–π –∫–ª–∏–µ–Ω—Ç –∏–∑ `src/client_giga.py`.

```python
%load_ext autoreload
%autoreload 2

from src.client_giga import get_gigachat_client
```

## 1. –ü–µ—Ä–≤—ã–π –∑–∞–ø—Ä–æ—Å: –æ–±—ä—è—Å–Ω–µ–Ω–∏–µ –ò–ò

```python
prompt_1 = "–ü—Ä–∏–≤–µ—Ç! –û–±—ä—è—Å–Ω–∏, —á—Ç–æ —Ç–∞–∫–æ–µ –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã–π –∏–Ω—Ç–µ–ª–ª–µ–∫—Ç –ø—Ä–æ—Å—Ç—ã–º–∏ —Å–ª–æ–≤–∞–º–∏."

with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    response_1 = giga.chat(prompt_1).choices[0].message.content

print("**–ü—Ä–æ–º–ø—Ç:**", prompt_1)
print("\n**–û—Ç–≤–µ—Ç –º–æ–¥–µ–ª–∏:**")
print(response_1)
```

## 2. –¢—Ä–∏ —Ç–∏–ø–∞ –∑–∞–¥–∞—á

```python
# 2a. –û–±—ä—è—Å–Ω–µ–Ω–∏–µ —Ç–µ—Ä–º–∏–Ω–∞
prompt_term = "–ß—Ç–æ —Ç–∞–∫–æ–µ —ç–º–±–µ–¥–¥–∏–Ω–≥–∏ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –Ω–µ–π—Ä–æ—Å–µ—Ç–µ–π? –û–±—ä—è—Å–Ω–∏ —Ç–∞–∫, —á—Ç–æ–±—ã –ø–æ–Ω—è–ª —Å—Ç–∞—Ä—à–µ–∫–ª–∞—Å—Å–Ω–∏–∫."

# 2b. –†–µ–∑—é–º–∏—Ä–æ–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞ (–≤–∑—è—Ç –∞–±–∑–∞—Ü –∏–∑ –í–∏–∫–∏–ø–µ–¥–∏–∏)
wiki_text = """
–ë–æ–ª—å—à–∏–µ —è–∑—ã–∫–æ–≤—ã–µ –º–æ–¥–µ–ª–∏ (LLM) ‚Äî —ç—Ç–æ —Ç–∏–ø –∏—Å–∫—É—Å—Å—Ç–≤–µ–Ω–Ω—ã—Ö –Ω–µ–π—Ä–æ–Ω–Ω—ã—Ö —Å–µ—Ç–µ–π, –æ–±—É—á–µ–Ω–Ω—ã—Ö –Ω–∞ –æ–≥—Ä–æ–º–Ω—ã—Ö –º–∞—Å—Å–∏–≤–∞—Ö —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö. 
–û–Ω–∏ —Å–ø–æ—Å–æ–±–Ω—ã –≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞—Ç—å —Å–≤—è–∑–Ω—ã–π —Ç–µ–∫—Å—Ç, –æ—Ç–≤–µ—á–∞—Ç—å –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã, –ø–µ—Ä–µ–≤–æ–¥–∏—Ç—å –∏ –≤—ã–ø–æ–ª–Ω—è—Ç—å –¥—Ä—É–≥–∏–µ —è–∑—ã–∫–æ–≤—ã–µ –∑–∞–¥–∞—á–∏. 
–û—Å–Ω–æ–≤—É –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–∞ —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—ã—Ö LLM —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä, –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–Ω–∞—è –≤ 2017 –≥–æ–¥—É.
"""
prompt_summary = f"–ö—Ä–∞—Ç–∫–æ (1‚Äì2 –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è) –ø–µ—Ä–µ—Å–∫–∞–∂–∏ —Å–ª–µ–¥—É—é—â–∏–π —Ç–µ–∫—Å—Ç:\n\n{wiki_text}"

# 2c. –û—Ç–≤–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å –ø–æ —Ç–µ–∫—Å—Ç—É
prompt_qa = f"–ù–∞ –æ—Å–Ω–æ–≤–µ —Å–ª–µ–¥—É—é—â–µ–≥–æ —Ç–µ–∫—Å—Ç–∞ –æ—Ç–≤–µ—Ç—å –Ω–∞ –≤–æ–ø—Ä–æ—Å: –ö–∞–∫–∞—è –∞—Ä—Ö–∏—Ç–µ–∫—Ç—É—Ä–∞ –ª–µ–∂–∏—Ç –≤ –æ—Å–Ω–æ–≤–µ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–∞ LLM?\n\n–¢–µ–∫—Å—Ç:\n{wiki_text}"

prompts = {
    "–û–±—ä—è—Å–Ω–µ–Ω–∏–µ —Ç–µ—Ä–º–∏–Ω–∞": prompt_term,
    "–†–µ–∑—é–º–∏—Ä–æ–≤–∞–Ω–∏–µ": prompt_summary,
    "–û—Ç–≤–µ—Ç –Ω–∞ –≤–æ–ø—Ä–æ—Å": prompt_qa
}

responses = {}
with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    for name, prompt in prompts.items():
        resp = giga.chat(prompt).choices[0].message.content
        responses[name] = resp
        print(f"\n### {name}")
        print(f"**–ü—Ä–æ–º–ø—Ç:** {prompt[:80]}..." if len(prompt) > 80 else f"**–ü—Ä–æ–º–ø—Ç:** {prompt}")
        print(f"\n**–û—Ç–≤–µ—Ç:**\n{resp}\n")
```

```python
prompt_hallucination = "–ö—Ç–æ –∏–∑–æ–±—Ä—ë–ª –ø–µ—Ä–≤—ã–π —á–∞—Ç-–±–æ—Ç –≤ 1945 –≥–æ–¥—É? –ù–∞–∑–æ–≤–∏ –∏–º—è –∏ —Å—Ç—Ä–∞–Ω—É."

with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    resp_hall = giga.chat(prompt_hallucination).choices[0].message.content

print("**–ü—Ä–æ–º–ø—Ç (–ø—Ä–æ–≤–æ–∫–∞—Ü–∏–æ–Ω–Ω—ã–π):**", prompt_hallucination)
print("\n**–û—Ç–≤–µ—Ç –º–æ–¥–µ–ª–∏:**")
print(resp_hall)
print("\n**–ê–Ω–∞–ª–∏–∑:**")
print("> –ü–µ—Ä–≤—ã–π —á–∞—Ç-–±–æ—Ç (ELIZA) –±—ã–ª —Å–æ–∑–¥–∞–Ω –≤ 1966 –≥–æ–¥—É –î–∂–æ–∑–µ—Ñ–æ–º –í–µ–π—Ü–µ–Ω–±–∞—É–º–æ–º –≤ –°–®–ê. –ú–æ–¥–µ–ª—å –≤—ã–¥–∞–ª–∞ –≤—ã–º—ã—à–ª–µ–Ω–Ω—ã–π —Ñ–∞–∫—Ç ‚Äî —ç—Ç–æ –≥–∞–ª–ª—é—Ü–∏–Ω–∞—Ü–∏—è.")
```
