Выполняем все 7 шагов для двух моделей: **Gigachat-2-Pro** и **Gigachat-2-Max**.

Обязательные приёмы:

- Роли/тон («мягко» vs «прямо»)
- Метки фактов: `[проверить]`, `[Тьюринг]`
- Формат: таблица Markdown перед связкой

## Шаг 1: Вступление

```python
%load_ext autoreload
%autoreload 2

from src.client_giga import get_gigachat_client
from src.utils import compare_responses, build_essay
TOPIC = "Проверка фактов: почему “умный ответ” не всегда верный"

prompt_intro = f"""Напиши краткое вступление (2-3 предложения) к эссе на тему «{TOPIC}». Аудитория: старшеклассники. Избегай выводов и оценок."""

with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    resp_pro = giga.chat(prompt_intro).choices[0].message.content

with get_gigachat_client(model="Gigachat-2-Max") as giga:
    resp_max = giga.chat(prompt_intro).choices[0].message.content

print(compare_responses(prompt_intro, resp_pro, resp_max, "Шаг 1: Вступление"))
```

## Шаг 2: Позиция (мягко и прямо)

```python
# Вариант А: мягко
prompt_soft = f"""Напиши абзац (3–4 предложения) с ненавязчивой авторской позицией по теме «{TOPIC}»: подводи к мысли через наблюдение/образ, не утверждай жёстко."""

# Вариант Б: прямо
prompt_direct = f"""Напиши абзац (3–4 предложения) с чёткой авторской позицией по теме «{TOPIC}». Начни со слов: «Я считаю, что …»."""

# Пример для одной модели (повтори для обеих при необходимости)
with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    soft_pro = giga.chat(prompt_soft).choices[0].message.content
    direct_pro = giga.chat(prompt_direct).choices[0].message.content

with get_gigachat_client(model="Gigachat-2-Max") as giga:
    soft_max = giga.chat(prompt_soft).choices[0].message.content
    direct_max = giga.chat(prompt_direct).choices[0].message.content

print("=== Мягкая позиция ===")
print(compare_responses(prompt_soft, soft_pro, soft_max, "Шаг 2A: Мягкая позиция"))

print("=== Прямая позиция ===")
print(compare_responses(prompt_direct, direct_pro, direct_max, "Шаг 2B: Прямая позиция"))
```

## Шаг 3: Пример 1 (школьная жизнь) — Zero-shot

```python
prompt_ex1_zs = f"""Напиши абзац (4–6 предложений) комментария к теме «{TOPIC}». Включи один конкретный пример из школьной жизни и поясни, что этот пример показывает. Избегай вымышленных цитат и ссылок.

Любые конкретные факты (годы, фамилии, названия) помечай в квадратных скобках: [1956], [Тьюринг]. Если не уверен — ставь [проверить]."""

with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    ex1_pro_zs = giga.chat(prompt_ex1_zs).choices[0].message.content

with get_gigachat_client(model="Gigachat-2-Max") as giga:
    ex1_max_zs = giga.chat(prompt_ex1_zs).choices[0].message.content

print(compare_responses(prompt_ex1_zs, ex1_pro_zs, ex1_max_zs, "Шаг 3: Пример 1 (zero-shot)"))
```

## Шаг 3: Пример 1 — Few-shot (с образцом)

```python
prompt_ex1_fs = f"""Пример для ориентира: «Учитель информатики показывает, как проверять факты ответов ИИ». Теперь придумай свой пример по аналогии.

Напиши абзац (4–6 предложений) комментария к теме «{TOPIC}». Включи один конкретный пример из школьной жизни и поясни, что этот пример показывает. Избегай вымышленных цитат и ссылок.

Любые конкретные факты (годы, фамилии, названия) помечай в квадратных скобках: [1956], [Тьюринг]. Если не уверен — ставь [проверить]."""

with get_gigachat_client(model="Gigachat-2-Pro") as giga:
    ex1_pro_fs = giga.chat(prompt_ex1_fs).choices[0].message.content

with get_gigachat_client(model="Gigachat-2-Max") as giga:
    ex1_max_fs = giga.chat(prompt_ex1_fs).choices[0].message.content

print(compare_responses(prompt_ex1_fs, ex1_pro_fs, ex1_max_fs, "Шаг 3: Пример 1 (few-shot)"))
```
